```
import numpy as np
import pandas as pd
```

```
df=pd.read_csv('RUUKPK.csv',delimiter=";", encoding='utf-8')
df.drop(['user_id','tweet_id','fullname','tweet_url','timestamp','timestamp_epochs','replies','retweets','likes','is_retweet','retweeter_username','retweeter_userid','retweet_id','html'],axis=1,inplace=True)

setuju = {'setuju':'setuju','menguatkan':'setuju','#SayaBersamaJokowi':'setuju','#DukungrevisiUUKPK':'setuju','dukung':'setuju','memperkuat':'setuju','#SayaBersamaJokowi':'setuju','menyetujui':'setuju','#RevisiUUKPKForNKRI':'setuju','anarkis':'setuju', 'perbaikan':'setuju','keharusan':'setuju'}
tolak = {'tolak':'tolak','#RezimMatiRasa':'tolak','menolak':'tolak','#judicialreview':'tolak','#KPKDijadikanBoneka':'tolak'}

def iya(k):
  x = (i for i in setuju if i in k)
  return '|'.join(map(setuju.get, x))
def tidak(k):
  x = (i for i in tolak if i in k)
  return '|'.join(map(tolak.get, x))

jsetuju = df['text'].map(iya)
jtolak = df['text'].map(tidak)

df['setuju'] = jsetuju
df['tolak'] = jtolak
df.head(1000)
```

```
df=pd.read_csv('RUUKPK.csv',delimiter=";", encoding='utf-8')
df.drop(['user_id','tweet_id','tweet_url','timestamp','timestamp_epochs','replies','retweets','likes','is_retweet','retweeter_username','retweeter_userid','retweet_id','html'],axis=1,inplace=True)

setuju = ['setuju','menguatkan','#SayaBersamaJokowi','#DukungrevisiUUKPK','dukung','memperkuat','#SayaBersamaJokowi','menyetujui','#RevisiUUKPKForNKRI','anarkis', 'perbaikan','keharusan']

pattern = '|'.join(setuju)     # joining list for comparision

Dukung=df['text'].str.contains(pattern)

n = df['username'].count()

df['Sikap'] = Dukung.replace((True,False), ('Setuju','Menolak'))
print(df.head(n))
print(1000*'==')
s = (df.Sikap == 'Setuju').sum()
t = (df.Sikap == 'Menolak').sum()
print("Jumlah Setuju :",s)
print("Jumlah Menolak :",t)
```
